```python
# %% [markdown]
# ## Step 0. Extract and Load Data
# 
# First, we would extract the tar file (if not already done) and load one of the expression files.
# In our example we assume the tar archive contains a file "GSE5859_expression.txt" (a tab‚Äêdelimited file)
# and a metadata file "GSE5859_metadata.txt" that contains cell-line covariates.
#
# (If using your own files, adjust the filenames and extraction paths accordingly.)

import tarfile
import os
import pandas as pd

# Define path to the tar file (adjust path as needed)
tar_path = "GSE5859_RAW.tar"
extract_path = "./GSE5859_extracted"

# Extract the tar archive if not already extracted
if not os.path.exists(extract_path):
    with tarfile.open(tar_path, "r") as tar:
        tar.extractall(path=extract_path)
        
# Load expression data and metadata.
# (For demonstration, we simulate these dataframes. Replace these lines with the actual file reading.)
# For example:
# expr_df = pd.read_csv(os.path.join(extract_path, "GSE5859_expression.txt"), sep="\t", index_col=0)
# meta_df = pd.read_csv(os.path.join(extract_path, "GSE5859_metadata.txt"), sep="\t")
import numpy as np

# Simulate an expression dataframe: rows are genes, columns are cell lines (n=60 cell lines)
np.random.seed(42)
genes = [f"Gene{i}" for i in range(1, 1001)]  # 1000 genes
cell_lines = [f"CL_{i}" for i in range(1, 61)]   # 60 cell lines
expr_data = np.abs(np.random.randn(1000, 60)) * 100  # simulated expression (TPM-like)
expr_df = pd.DataFrame(expr_data, index=genes, columns=cell_lines)

# Simulate metadata (covariates) for the cell lines
meta_df = pd.DataFrame({
    "CellLine": cell_lines,
    "TissueType": np.random.choice(["Breast", "Lung", "Colon", "Melanoma"], size=60),
    "Age": np.random.randint(20, 80, size=60),
    "Batch": np.random.choice(["Batch1", "Batch2"], size=60)
})
meta_df.set_index("CellLine", inplace=True)

print("Expression data shape:", expr_df.shape)
print("Metadata shape:", meta_df.shape)
```

```python
# %% [markdown]
# ## Step 1. Feature Selection and Engineering
# 
# In this example, we simulate a scenario in which we want to predict the expression of a particular gene 
# (say, "Gene10") using available cell-line covariates. In practice, you might have additional features 
# such as genetic variants (e.g. eQTLs) or technical covariates (e.g. RIN scores).
#
# We will first merge the covariate information (from `meta_df`) with the expression value for Gene10.
# Then, we will perform a simple transformation (e.g. standardization) and also apply a dimensionality 
# reduction technique such as PCA if you have many predictors.

from sklearn.preprocessing import StandardScaler
from sklearn.decomposition import PCA
import matplotlib.pyplot as plt
import seaborn as sns

# Let Y be the expression of Gene10 (for demonstration)
Y = expr_df.loc["Gene10"]
# Merge Y with covariates; each cell line becomes an observation.
data = meta_df.copy()
data["Gene10_expr"] = Y
data.reset_index(inplace=True)

# For categorical covariates like TissueType and Batch, create dummy variables
data_encoded = pd.get_dummies(data, columns=["TissueType", "Batch"], drop_first=True)

print("Data after encoding:")
print(data_encoded.head())

# Separate predictors (X) and response (Y)
X = data_encoded.drop(columns=["CellLine", "Gene10_expr"])
y = data_encoded["Gene10_expr"]

# Standardize predictors
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)

# If you have many predictors, you might perform PCA. Here, we simply run PCA and plot the first two PCs.
pca = PCA(n_components=2)
X_pca = pca.fit_transform(X_scaled)

plt.figure(figsize=(6,4))
sns.scatterplot(x=X_pca[:,0], y=X_pca[:,1], hue=data_encoded["TissueType_Breast"] if "TissueType_Breast" in data_encoded.columns else None)
plt.xlabel("PC1")
plt.ylabel("PC2")
plt.title("PCA of Cell-Line Covariates")
plt.show()

# Equation (displayed using LaTeX in matplotlib):
plt.figure(figsize=(6, 1))
plt.text(0.1, 0.5, r"Standardization: $X_{\mathrm{scaled}} = \frac{X - \mu_X}{\sigma_X}$", fontsize=14)
plt.axis('off')
plt.show()
```

```python
# %% [markdown]
# ## Step 2. Model Choice
# 
# Here we select a gradient-boosted tree model (using XGBoost) to predict gene expression from our covariates.
# We use mean squared error (MSE) as the loss function. Later, we will also compute a measure of uncertainty.
#
# We demonstrate the training using cross-validation.

import xgboost as xgb
from sklearn.model_selection import train_test_split, KFold, cross_val_score

# Split data into training and validation sets.
# In a real scenario, you might set aside a subset as your "gold-standard" and use the remainder as unlabeled.
X_train, X_val, y_train, y_val = train_test_split(X_scaled, y, test_size=0.3, random_state=42)

# Define the XGBoost regressor
xgb_reg = xgb.XGBRegressor(objective='reg:squarederror', n_estimators=100, learning_rate=0.1, max_depth=3, random_state=42)

# Perform 5-fold cross-validation on the training set
cv = KFold(n_splits=5, shuffle=True, random_state=42)
cv_scores = cross_val_score(xgb_reg, X_train, y_train, cv=cv, scoring='neg_mean_squared_error')
rmse_scores = np.sqrt(-cv_scores)
print("Cross-validated RMSE:", rmse_scores.mean())

# Fit the model on the full training data
xgb_reg.fit(X_train, y_train)

# Plot the learned relationship (predicted vs true) on the validation set:
y_pred = xgb_reg.predict(X_val)
plt.figure(figsize=(6,4))
plt.scatter(y_val, y_pred, color='blue', edgecolor='k')
plt.xlabel("True Expression")
plt.ylabel("Predicted Expression")
plt.title("XGBoost Predictions on Validation Data")
plt.plot([y_val.min(), y_val.max()], [y_val.min(), y_val.max()], 'r--')  # identity line
plt.show()

# Display the loss function equation used (MSE)
plt.figure(figsize=(6, 1))
plt.text(0.1, 0.5, r"MSE Loss: $\mathrm{MSE} = \frac{1}{n}\sum_{i=1}^{n}(y_i - \hat{y}_i)^2$", fontsize=14)
plt.axis('off')
plt.show()
```

```python
# %% [markdown]
# ## Step 3. Training Strategy and Generating Predictions
# 
# We now simulate the process of splitting the data into a "gold‚Äêstandard" set and an "unlabeled" set.
# For demonstration, let‚Äôs assume 30% of the 60 cell lines are gold‚Äêstandard (with true measurements) and 70%
# serve as the unlabeled set. We then use our trained model to predict the gene expression on both sets.
#
# (In practice, you might use additional data for the unlabeled portion.)

# Split the full dataset (here, using our small 60-cell-line dataset) into gold-standard and unlabeled:
gold_idx, unlabeled_idx = train_test_split(np.arange(len(y)), test_size=0.7, random_state=42)
print("Gold-standard indices:", gold_idx)
print("Unlabeled indices:", unlabeled_idx)

# For gold-standard, we have true y and predictors X_scaled; for unlabeled, we only use X_scaled.
X_gold = X_scaled[gold_idx]
y_gold = y.iloc[gold_idx].values
X_unlabeled = X_scaled[unlabeled_idx]

# Generate predictions on both sets using our already-trained model:
pred_gold = xgb_reg.predict(X_gold)
pred_unlabeled = xgb_reg.predict(X_unlabeled)

# Plot the distribution of predictions vs. true values on the gold-standard set:
plt.figure(figsize=(6,4))
sns.histplot(pred_gold, color='skyblue', label="Predictions", kde=True)
sns.histplot(y_gold, color='salmon', label="True Values", kde=True)
plt.xlabel("Gene10 Expression")
plt.title("Distribution on Gold-Standard Set")
plt.legend()
plt.show()
```

```python
# %% [markdown]
# ## Step 4. Integration into Prediction-Powered Inference
# 
# Now, we follow the prediction-powered inference idea.
# 
# **a. Compute the Imputed Estimator:**  
# For example, suppose we wish to estimate the mean expression level of Gene10 in the unlabeled data.
# We first compute the mean of the predictions over the unlabeled set.
# 
# **b. Compute the Empirical Rectifier:**  
# Using the gold‚Äêstandard set, we calculate the average bias (i.e. the mean difference between the predictions
# and the true expression).
#
# **c. Combine to Form the Prediction-Powered Estimator:**  
# The final estimator is:
#
# \[
# \hat{\theta}_{PP} = \frac{1}{N}\sum_{i \in \text{unlabeled}} f(X_i) - \hat{\Delta},
# \]
#
# where
#
# \[
# \hat{\Delta} = \frac{1}{n}\sum_{i \in \text{gold-standard}} \left( f(X_i) - Y_i \right).
# \]
#
# **d. Confidence Interval:**  
# For simplicity, we will compute the standard error from both parts and form a 95% CI.
#
# (In practice, you might use a more advanced method as outlined in the prediction-powered inference algorithms.)

N = len(pred_unlabeled)
n_gold = len(pred_gold)

theta_imputed = np.mean(pred_unlabeled)
delta_hat = np.mean(pred_gold - y_gold)
theta_PP = theta_imputed - delta_hat

print("Imputed estimator (mean on unlabeled):", theta_imputed)
print("Empirical rectifier (bias estimate):", delta_hat)
print("Prediction-powered estimator:", theta_PP)

# Compute variance estimates:
var_unlabeled = np.var(pred_unlabeled, ddof=1)
var_gold = np.var(pred_gold - y_gold, ddof=1)

# Standard errors:
se_unlabeled = np.sqrt(var_unlabeled / N)
se_gold = np.sqrt(var_gold / n_gold)

# Combined standard error (assuming independence):
se_total = np.sqrt(se_unlabeled**2 + se_gold**2)

# 95% Confidence Interval:
z = 1.96  # for 95% CI
CI_lower = theta_PP - z * se_total
CI_upper = theta_PP + z * se_total

print("95% Confidence Interval for the mean expression:", (CI_lower, CI_upper))

# Plot the CI with the estimator:
plt.figure(figsize=(6, 1))
plt.errorbar(1, theta_PP, yerr=z*se_total, fmt='o', color='green', capsize=5)
plt.xlim(0.5, 1.5)
plt.xticks([])
plt.ylabel("Gene10 Expression")
plt.title("Prediction-Powered Estimator with 95% CI")
plt.show()

# Display the corresponding equation (using LaTeX):
plt.figure(figsize=(6, 1))
plt.text(0.1, 0.5, 
         r"$\hat{\theta}_{PP} = \frac{1}{N}\sum_{i \in \text{unlabeled}} f(X_i) - \frac{1}{n}\sum_{i \in \text{gold}} (f(X_i) - Y_i)$", 
         fontsize=14)
plt.axis('off')
plt.show()
```
